```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("dataset2.csv")
df.head()
```

```python
# --------------------------
# 1. Rat arrivals over time (hours after sunset)
# --------------------------
plt.figure(figsize=(6,4))
sns.lineplot(x="hours_after_sunset", y="rat_arrival_number", data=df, marker="o")
plt.title("Rat Arrivals Over Time After Sunset")
plt.xlabel("Hours After Sunset")
plt.ylabel("Number of Rat Arrivals")
plt.show()
```

```python
#  --------------------------
# 2. Bat landings vs rat arrivals
# --------------------------
plt.figure(figsize=(6,4))
sns.scatterplot(x="rat_arrival_number", y="bat_landing_number", data=df)
plt.title("Bat Landings vs Rat Arrivals")
plt.xlabel("Rat Arrival Number")
plt.ylabel("Bat Landing Number")
plt.show()
```

```python
# --------------------------
# 3. Rat minutes vs food availability
# --------------------------
plt.figure(figsize=(6,4))
sns.scatterplot(x="rat_minutes", y="food_availability", data=df)
plt.title("Food Availability vs Rat Presence Duration")
plt.xlabel("Rat Minutes (total duration rats present)")
plt.ylabel("Food Availability")
plt.show()
```

```python
# --------------------------
# 4. Bat landings vs food availability
# --------------------------
plt.figure(figsize=(6,4))
sns.scatterplot(x="food_availability", y="bat_landing_number", data=df)
plt.title("Bat Landings vs Food Availability")
plt.xlabel("Food Availability")
plt.ylabel("Bat Landing Number")
plt.show()
```
